```python
import pandas as pd
import glob
```

```python
#ETL for tripdata
```

```python
riderfiles_2017_2018 = glob.glob("JC*.csv")
```

```python
list_ =[]
for file_ in riderfiles_2017_2018:
    rider_df = pd.read_csv(file_,index_col=None, header=0)
    list_.append(rider_df)

riders = pd.concat(list_, axis = 0, ignore_index = True)
```

```python
riders
```

```python
riders['gender'] = riders['gender'].map({1: 'male', 2: 'female', 0: 'not reported'})
```

```python
riders
```

```python
riders.to_csv("riders_merged.csv")
```
