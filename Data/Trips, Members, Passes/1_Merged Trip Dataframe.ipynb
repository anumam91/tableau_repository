```python
import pandas as pd
import glob
```

```python
allfiles = glob.glob("*_Q4.csv")
print(allfiles)
```

```python
list_ = []

for file_ in allfiles:
    df = pd.read_csv(file_,index_col=None, header=0)
    list_.append(df)

frame = pd.concat(list_, axis = 0, ignore_index = True)
frame.to_csv("allyears_merged.csv")
```
